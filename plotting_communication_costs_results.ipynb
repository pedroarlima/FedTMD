```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df_without_selection = pd.read_csv('./aggregated_communication_costs_results_without_selection.csv')
df_with_selection = pd.read_csv('./aggregated_communication_costs_results_with_selection.csv')


```

```python
# Agrupar por 'num_rounds' de 'cost' e 'bound'
df_with_selection.groupby('num_rounds')['cost']
df_without_selection.groupby('num_rounds')['cost']

# Obter os valores únicos de 'num_rounds'
rounds = df_with_selection['num_rounds'].unique()

# Ajustar a largura das barras para que elas fiquem lado a lado
bar_width = 0.35

# Calcular a posição das barras
index = np.arange(len(rounds))

# Criar as barras lado a lado
plt.bar(index - bar_width/2, df_with_selection['cost'], bar_width, label='Cost with Selection', color='blue')
plt.bar(index + bar_width/2, df_with_selection['bound'], bar_width, label='Bound without Selection', color='orange')

# Título e rótulos do eixo x e y
plt.title('Communication Costs and Bound')
plt.xlabel('Total Rounds')
plt.ylabel('Cost')

# Ajuste o layout do gráfico
plt.tight_layout()

# Configurar os rótulos do eixo x para serem os valores de 'num_rounds'
plt.xticks(index, rounds)

# Adicionar uma grade
plt.grid()

# Adicionar a legenda
plt.legend()

# Mostrar o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Ordenar os valores únicos de 'num_rounds' em ordem crescente
rounds = df_with_selection['num_rounds'].unique()

# Ajustar a largura das barras para que elas fiquem lado a lado
bar_width = 0.25  # Reduzindo a largura para acomodar três barras lado a lado

# Calcular a posição das barras
index = np.arange(len(rounds))

# Criar as barras lado a lado
plt.bar(index - bar_width, df_with_selection['cost'], bar_width, label='Cost with Selection', color='blue')
plt.bar(index, df_without_selection['cost'], bar_width, label='Cost without Selection', color='orange')
plt.bar(index + bar_width, df_without_selection['bound'], bar_width, label='Communication Bound', color='green')

# Configurar os rótulos do eixo x para serem os valores de 'num_rounds' ordenados
plt.xticks(index, rounds)

# Título e rótulos do eixo x e y
plt.title('Communication Costs x Bound')
plt.xlabel('Total Rounds')
plt.ylabel('Costs (bytes)')

# Ajuste o layout do gráfico
plt.tight_layout()

# Adicionar uma grade
plt.grid()

# Adicionar a legenda
plt.legend()

# Mostrar o gráfico
plt.show()
```
